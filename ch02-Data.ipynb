**Attribute Types**
- Categorical : 유한개의 값 <- Decision Tree
  - Nominal : 그냥 구분되는 할당에 불과, ID 등, =, not=
  - ordinal : 순서를 갖는, <, >
- Numerical  : value <- SVM 등
  - Interval : 정수 구간, +, -
  - Ratio : 실수, x,/

##  Dimensionality Reduction

>*고차원 데이터의 내재된 intrinsic dimension을 찾는 방법

>*고차원에서는 우리가 알고 있는 distance measure의 결과가 다르게 동작할 수 있다. 차원이 높을수록 random하게 분포시킨 점들 사이의 min distance와 max distance의 차이가 작아진다.


>*PCA, Sigular value decomposition, Eigen value decomposition techniques 등이 사용됨.

>### -Dimensionality Reduction
>> - Feature selection

>>> : attribute 중 일부만 취한다. Target과의 correlation을 계산해서 낮은 attributes는 제외한다.

>>> ex) Decision Tree에서 selection을 통해 결과가 좋아지는지를 확인하는 방식으로 experiment 한다.

>> - Feature extration

>>> : 가공을 통해 줄인다. PCA, singular, eigen decomposition 등

##  Discretization

##  Attribute Transformation

>- Standardization : (X-mean)/var

>- Normalization : 최소 - 최대 구간을 0 - 1 사이로 mapping, attribute간 scale이 크게 차이가 나는 경우에는 큰 값을 갖는 attribute에 의해 dominent하게 measure가 작동할 수 있으므로 normalization해서 attribute간 scale을 맞춰준다.
